```python
import seaborn as sns
import matplotlib.pyplot as plt
# to check how the datapoints are distributed if plotted against "charges"
sns.histplot(insurance["charges"], kde=True)
plt.show()
```

The above graph depects data is normally skewed to the left. which indicates that inorder to normalize this curve to correct centralized distribution, we have to apply log function

```python
insurance["charges_log"]=np.log(insurance["charges"])
sns.histplot(insurance["charges_log"],kde=True,color="red")
plt.title("log transformed with charges")
```

```python
import matplotlib.pyplot as plt

for col in ["sex", "smoker", "region"]:
    insurance[col].value_counts().plot(kind="pie", autopct="%.1f%%")
    plt.title(f"proportion of {col}")
    plt.ylabel("")
    plt.show()
```

```python
import matplotlib.pyplot as plt

colors = insurance["smoker"].map({"yes": "red", "no": "blue"})

plt.scatter(insurance["age"], insurance["charges_log"], c=colors, alpha=0.4)
plt.xlabel("age")
plt.ylabel("charges_log")
plt.title("age vs charges by smoker status")
plt.show()
```

```python
insurance = pd.get_dummies(insurance, columns=["sex", "smoker", "region"], drop_first=True)
print(insurance.columns)
```
